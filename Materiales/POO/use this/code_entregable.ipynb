```python
import random
from IPython.display import clear_output



import random
## superclass ##
## PLAYER WARRIOR ##   
# Superclase Guerrero
#La super clase guerrero tiene varios atributos sin "definir". 
class Warrior:
    def __init__(self, name, HP, ATK, DEF, SPD):
        self.nombre = name
        self.salud = HP
        self.fuerza = ATK
        self.defensa = DEF
        self.velocidad = SPD

    def atacar(self, objetivo):
        pass  # Será implementado en las subclases

    def defender(self, daño):
        # Método común para reducir el daño basado en la defensa
        daño_recibido = max(0, daño - (self.defensa * 0.4))  # Ajustable en subclases
        self.salud -= daño_recibido
        return daño_recibido

    def sigue_vivo(self):
        return self.salud > 0

    def mostrar_estadisticas(self):
        print(f"Nombre: {self.nombre}")
        print(f"Salud: {self.salud}")
        print(f"Fuerza: {self.fuerza}")
        print(f"Defensa: {self.defensa}")
        print(f"Velocidad: {self.velocidad}")
        
# ------------------------------------------------------------------------------------------------------------------------------------------ #

     

class Knight(Warrior):
    warrior_name  = 'Matthew'     
    HP = random.randint(100,151)
    ATK = random.randint(30,41)
    DEF = random.randint(50,71)
    SPD = random.randint(10,21)
    
    ## you must pass the variable you have put Knight into when executing the obeject ##
        # e.g #
    # atak = Knight
    # atak.atk(atak)
    def attack(self,enemy):
        self.enemy = enemy
        
        
        damage = self.ATK * 1.2## attack ##
        
        enemy.DEF(damage)
        
    ## you must pass the variable you have put Knight into when executing the obeject ##
    # same here #

    def defense(self,enemy):
        
        self.enemy = enemy
        
        damage_taken = enemy.ATK - (self.DEF * 0.6)
        
        self.HP = self.HP - damage_taken
    
    # ------------------------------------------------------------------------------------------------------------------------------------------ #
class Archer(Warrior):
    warrior_name  = 'Chris'
    HP = random.randint(90,121)
    ATK = random.randint(25,36)
    DEF = random.randint(10,21)
    SPD = random.randint(40,61)
    
    ## ATTACK ##
    def attack(self, enemy):
        
        self.enemy = enemy
        ## 30% chance to perform an attack ##
        
        chance_to_attack = random.randint(0,101)
        chance = True
        
        while chance == True:
            if chance_to_attack > 30:## 30% to land ##
                normal_attack = self.ATK * 1 ## damage of attack ##
                enemy.DEF(enemy.DEF - normal_attack)
                # print(f'The arrow hit! damage({damage})')
                break
        
            if chance_to_attack <= 30:
                super_damage = self.ATK * 1
                super_def_Break = enemy.DEF / 2
                ## ignores 50% of the opponent's defense ##
                enemy.DEF(super_def_Break)
                enemy.HP(super_damage)
                # print(f'SUPER HIT! enemy defense halfed! damage({special_attack_damage})  enemy defense ({enemy_defense}))')
                break 
        
    def defense(self,enemy):
        self.enemy = enemy
        
        life = True
        
        while life == True:
            ## attack dodge ##
            chance_to_evade = random.randint(0,100)
            
            if chance_to_evade <= 20:
                # print('you dodged the attack!')
                break
            
            
            ## if damage lands ##
            damage_taken = enemy.ATK - (self.DEF * 0.4)
            damage_taken = round(damage_taken,2)
            
            self.HP = self.HP - damage_taken
            
            life = False
            break        
        
        
        
        
        
        
    
# ------------------------------------------------------------------------------------------------------------------------------------------ #
class Mage(Warrior):
    warrior_name  = 'Dan'
    HP = random.randint(50,81)
    ATK = random.randint(40,61)
    DEF = random.randint(10,21)
    SPD = random.randint(20,31)
    
    
    def attack(self,enemy):
        self.enemy = enemy
        
        
        
    
    
    
# ------------------------------------------------------------------------------------------------------------------------------------------ #
class Assassin(Warrior):
    warrior_name  = 'Oscar'
    HP = random.randint(60,91)
    ATK = random.randint(35,46)
    DEF = random.randint(15,26)
    SPD = random.randint(50,71)  
    
# ------------------------------------------------------------------------------------------------------------------------------------------ #
class Healer(Warrior):
    warrior_name  = 'James'
    HP = random.randint(80,121)
    ATK = random.randint(10,21)
    DEF = random.randint(20,31)
    SPD = random.randint(30,41)
    
    
## --------------------ENEMY WARRIOR --------------------##     
    
class Knight_enemy(Warrior):
    warrior_name  = 'Taylor'     
    HP = random.randint(100,151)
    ATK = random.randint(30,41)
    DEF = random.randint(50,71)
    SPD = random.randint(10,21)
    
    
        ## you must pass the variable you have put Knight into when executing the obeject ##
        # e.g #
    # atak = Knight
    # atak.atk(atak)
    def attack(self,enemy):
        self.enemy = enemy
        
        enemy.hp - self.ATK * 1.2
        
        
    ## you must pass the variable you have put Knight into when executing the obeject ##
    # same here #

    def defense(self,enemy):
        
        self.enemy = enemy
        
        damage_taken = enemy.ATK - (self.DEF * 0.6)
        
        self.HP = self.HP - damage_taken
        
        
    
class Archer_enemy(Warrior):
    warrior_name  = 'Ethan'
    HP = random.randint(70,101)
    ATK = random.randint(25,36)
    DEF = random.randint(20,31)
    SPD = random.randint(40,61)
    
        ## ATTACK ##
    def attack(self, enemy):
        
        self.enemy = enemy
        ## 30% chance to perform an attack ##
        
        chance_to_attack = random.randint(0,101)
        chance = True
        
        while chance == True:
            if chance_to_attack > 30:
                damage = self.ATK * 1.0
                print(f'The arrow hit! damage({damage})')
                break
        
            if chance_to_attack <= 30:
                special_attack_damage = self.ATK * 1.0
                
                ## ignores 50% of the opponent's defense ##
                
                enemy_defense = enemy.DEF / 2
                print(f'SUPER HIT! enemy defense halfed! damage({special_attack_damage})  enemy defence ({enemy_defense}))')
                break 
            
            
            
    def defense(self,enemy):
        self.enemy = enemy
        
        life = True
        
        while life == True:
            ## if health goes to zero ##
            if self.HP < 0:
                life = False
                break
            
            ## attack dodge ##
            chance_to_evade = random.randint(0,100)
            
            if chance_to_evade <= 20:
                print('you dodged the attack!')
                break
            
            
            ## if damage lands ##
            damage_taken = enemy.ATK - (self.DEF * 0.4)
            damage_taken = round(damage_taken,2)
            
            self.HP = self.HP - damage_taken
            
            print(f'youve been hit damage taken({damage_taken} ARCHER HP ({self.HP}))')
            life = False
            break    
    
    
    
    
    
class Mage_enemy(Warrior):
    warrior_name  = 'Jack'
    HP = random.randint(50,81)
    ATK = random.randint(40,61)
    DEF = random.randint(10,21)
    SPD = random.randint(20,31)
    
class Assassin_enemy(Warrior):
    warrior_name  = 'Jac'
    HP = random.randint(60,91)
    ATK = random.randint(35,46)
    DEF = random.randint(15,26)
    SPD = random.randint(50,71)  
    
class Healer_enemy(Warrior):
    warrior_name  = 'Oliver'
    HP = random.randint(80,121)
    ATK = random.randint(10,21)
    DEF = random.randint(20,31)
    SPD = random.randint(30,41)
    
  
```
