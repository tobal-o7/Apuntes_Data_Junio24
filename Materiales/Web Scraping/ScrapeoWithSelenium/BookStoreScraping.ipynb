```python
from bs4 import BeautifulSoup as bs
import pandas as pd 
import requests as req
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager # Se encarga de descargar el driver de chrome y de mantenerlo actualizado
from selenium.webdriver.chrome.options import Options
```

```python
url = 'https://books.toscrape.com/?'
html = req.get(url).content
soup=bs(html, 'html.parser')
```

```python
for page in range(1,51):
    url = 'https://books.toscrape.com/catalogue/page-{page}.html'
    html = req.get(url).content
    soup=bs(html, 'html.parser')
    
    
    list_books= []
    #Listado de ofertas
    books = soup.find_all('article', {'class':'product_pod'})

    for i in books:
            
        dicc_books = {} #Diccionario donde almacenamos datos de 1 oferta
                
        book_name = i.find('h3', {'title':i}).text.strip()
        dicc_books['book_name'] = book_name
            
        product_price = i.find('p', {'class':'price_color'}).text.strip()
        dicc_books['product_price'] = product_price
                
        book_link = i.find('a'), {'href'}
        dicc_books['book_link'] = book_link
                
        # product_availability = i.find('i', {'class':i})
                
                
                
        #Appendeamos a nuestra lista de diccionarios de respuesta
        list_books.append(dicc_books) 
        print(list_books)
        print(dicc_books)
```
